# Split, separacion y preparacion de los datos

## Separacion datos de entrenamiento y datos de prueba

## OneHotEncoder, haciendo encoding en el set de datos para manejar el Machine Learning

```python
one_hot = make_column_transformer((OneHotEncoder(drop='if_binary'),
                                   ['contract_type','internet_service','payment_method']),
                                  remainder='passthrough',
                                  sparse_threshold=0,
                                  force_int_remainder_cols=False)
```

```python
X = one_hot.fit_transform(X)
```

```python
one_hot.get_feature_names_out(columnas)
```

```python
pd.DataFrame(X, columns=one_hot.get_feature_names_out(columnas))
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y, test_size=0.2,random_state=201)
```

# Modelos Machine Learning

## 1.- Modelo de Regresion Logistica

```python
modelo_log = LogisticRegression(max_iter=1000, class_weight='balanced')
modelo_log.fit(X_train, y_train)

y_pred_log = modelo_log.predict(X_test)

print("Regresion Logistica")
print(f"{classification_report(y_test, y_pred_log)}")
```

```python
cm_log = confusion_matrix(y_test, y_pred_log)

# Grafico
plt.figure()
plt.imshow(cm_log)
plt.title("Matriz de Confusion - Regresion logistica")
plt.xlabel("Prediccion")
plt.ylabel("Real")

for i in range(cm_log.shape[0]):
    for j in range(cm_log.shape[1]):
        plt.text(j, i, cm_log[i, j], ha="center", va="center")

plt.xticks([0,1], ["No", "Yes"])
plt.yticks([0,1], ["No", "Yes"])

plt.show()
```

## 2.- Modelo KNN (KNeighborsClassifier)

```python
scaler = StandardScaler()
X_entranemintoEscalada = scaler.fit_transform(X_train)
X_pruebaEscalada = scaler.fit_transform(X_test)
```

```python
knn_classifier = KNeighborsClassifier(n_neighbors=5)
```

```python
knn_classifier.fit(X_entranemintoEscalada, y_train)
```

```python
y_pred = knn_classifier.predict(X_pruebaEscalada)

print(f'Precision: {accuracy_score(y_test, y_pred)}')
print(f'{classification_report(y_test, y_pred)}')
```

```python
pipeline_knn = Pipeline([
    ('scaler', StandardScaler()),
    ('knn', KNeighborsClassifier(n_neighbors=5))
])
```

```python
pipeline_knn.fit(X_train, y_train)
```

```python
print(f'Precision: {accuracy_score(y_test, y_pred)}')
print(f'{classification_report(y_test, y_pred)}')
print(f'{confusion_matrix(y_test, y_pred)}')
```

```python
param_grid = {
    'knn__n_neighbors':range(1, 21)
}

grid_search = GridSearchCV(
    pipeline_knn,
    param_grid,
    cv=5,
    scoring='accuracy'
)

grid_search.fit(X_train, y_train)

print(f'El mejor k es: {grid_search.best_params_}')
print(f'El mejor score CV es: {grid_search.best_score_}')
```

```python
best_model = grid_search.best_estimator_
y_pred = best_model.predict(X_test)

print(f'Puntuacion final es: {accuracy_score(y_test, y_pred)}')
```

```python
print(classification_report(y_test, y_pred))
```

```python
cm_knn = confusion_matrix(y_test, y_pred)

# Grafico
plt.figure()
plt.imshow(cm_knn)
plt.title("Matriz de Confusion - KNN")
plt.xlabel("Prediccion")
plt.ylabel("Real")

for i in range(cm_knn.shape[0]):
    for j in range(cm_knn.shape[1]):
        plt.text(j, i, cm_knn[i, j], ha="center", va="center")

plt.xticks([0,1], ["No", "Yes"])
plt.yticks([0,1], ["No", "Yes"])

plt.show()
```

## 3.- Modelo Random Forest

```python
modelo_RF = RandomForestClassifier(
    n_estimators=200,
    random_state=201,
    class_weight='balanced'
)
```

```python
modelo_RF.fit(X_train, y_train)
```

```python
y_pred_rf = modelo_RF.predict(X_test)

print(f'Precision: {accuracy_score(y_test, y_pred_rf)}')
print(f'{classification_report(y_test, y_pred_rf)}')
print(f'{confusion_matrix(y_test, y_pred_rf)}')
```

```python
param_grid = {
    'n_estimators': [100, 200, 300],
    'max_depth': [None, 10, 20],
    'min_samples_split': [2, 5, 10, 20],
}

grid_rf = GridSearchCV(
    RandomForestClassifier(class_weight='balanced', random_state=42),
    param_grid,
    cv=5,
    scoring='recall'
)

grid_rf.fit(X_train, y_train)

print(f'Mejores parametros: {grid_rf.best_params_}')
```

```python
best_rf = grid_rf.best_estimator_
y_pred_rf = best_rf.predict(X_test)

print(f'Precision: {accuracy_score(y_test, y_pred_rf)}')
print(f'{classification_report(y_test, y_pred_rf)}')
print(f'{confusion_matrix(y_test, y_pred_rf)}')
```

```python
y_proba = modelo_RF.predict_proba(X_test)[:, 1]

import numpy as np

threshold = 0.35
y_pred_custom = (y_proba >= threshold).astype(int)

from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred_custom))
```

```python
from sklearn.metrics import average_precision_score

pr_auc = average_precision_score(y_test, y_proba)
print("PR-AUC:", pr_auc)
```
